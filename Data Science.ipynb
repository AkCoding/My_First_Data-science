# 1 .  Plot day by day stats of udacityâ€™s lesson views


```python
import pandas as pd 
from matplotlib import pyplot as plt
df = pd.read_csv("F:\\csvFile\\udacity_view_data.csv")
ldf = df[df['lesson'] == '1-0']
ldf = ldf.set_index('date')
plt.figure(figsize = (20,7))
plt.plot(ldf.views, label = 'no. of Views')
plt.xticks(rotation = 90)
plt.title('udacity lesson watched (daily data)',fontsize = 27)
plt.xlabel('Date', fontsize = 19)
plt.ylabel('Views', fontsize = 19)
plt.legend(loc = 4, fontsize = 14)
plt.grid(True)
plt.show()
```

# 2 . Animate the above plot from minimum date to maximum date

```python
import pandas as pd 
from matplotlib import pyplot as plt
data = pd.read_csv("F:\\csvFile\\udacity_view_data.csv")
df = data.groupby('date')['views'].sum()
df = df.reset_index()
x = df.loc[:,'date']
y = df.loc[:,'views']
plt.figure(figsize = (18, 6))
plt.bar(x, y)
plt.xticks(rotation = 90)
plt.title('Views Count', fontsize = 27)
```

# 3. Get which lesson is least and maxed viewed lesson

```python
import pandas as pd 
data = pd.read_csv("F:\\csvFile\\udacity_view_data.csv")
df = data.groupby('lesson')['views'].sum()
df = df.reset_index()
min_value = df[df['views'] == df['views'].min()]
max_value = df[df['views'] == df['views'].max()]
print('least viewed lesson \n', min_value, '\n\n\n Maxed viewed lesson \n', max_value)
```

# 5. Segment data based on its date (Ex. 14-09.csv, 15-09.csv) 

```python
import pandas as pd 
from datetime import datetime
df = pd.read_csv("F:\\csvFile\\udacity_view_data.csv")
for date in df['date'].unique():
#     print(date)
    date_time = datetime.strptime(date, '%Y/%m/%d %H:%M:%S')
    formatted = date_time.strftime("%d-%m")
    print(formatted)
    df[df['date'] == date].to_csv(formatted + str('.csv'))
    print('export data Successfully')
    
```
